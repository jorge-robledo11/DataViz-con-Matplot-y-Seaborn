```python
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.simplefilter("ignore")

x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
countries = ['India', 'Japón', 'México', 'Colombia', 'Alemania']
population = [1000, 800, 900, 700, 600]
```

```python
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(6,6))

ax.bar(countries, population, width=0.5, color=['olive', 'blue', 'pink', 'magenta', 'orange'])
ax.set_title('Países vs Población', fontsize=16)
ax.set_xlabel('Países', fontsize=14)
ax.set_ylabel('Cantidad Poblacional', fontsize=14)
ax.set_xticks(np.arange(5), countries, rotation=45)
ax.grid()
fig.tight_layout()
```

```python
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(6,6))

ax.barh(countries, population, height=0.5, color=['olive', 'blue', 'pink', 'magenta', 'orange'])
ax.set_title('Países vs Población', fontsize=16)
ax.set_ylabel('Países', fontsize=14)
ax.set_xlabel('Cantidad Poblacional', fontsize=14)
ax.set_yticks(np.arange(5), countries, rotation=45)
ax.grid()
fig.tight_layout()
```
