```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(1337)
data = np.random.randint(1, 50, 100)
data
```

```python
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(6,6))

ax.hist(data, color='orange', bins=10)
ax.set_title('Histograma', fontsize=16)
ax.grid()
fig.tight_layout()
```

```python
plt.style.use('seaborn-colorblind')
fig, ax = plt.subplots(figsize=(6,6))

ax.hist(data, color='white', bins=10, histtype='step')
ax.set_title('Histograma', fontsize=16)
ax.grid()
fig.tight_layout()
```
