```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter("ignore")

tips = sns.load_dataset('tips')
tips = tips.rename(columns={'sex': 'gender'})
tips = tips.rename(str.capitalize, axis='columns')
tips
```

```python
sns.countplot(data=tips, x='Day', hue='Gender')
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(6,6))
sns.stripplot(data=tips, x='Day', y='Tip', hue='Gender', dodge=True)
fig.tight_layout()
```

```python
fig = plt.figure(figsize=(6,6))
sns.swarmplot(data=tips, x='Day', y='Tip', hue='Gender', dodge=True)
fig.tight_layout()
```

```python
fig = plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='Day', y='Tip', hue='Gender', dodge=True)
fig.tight_layout()
```

```python
fig = plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='Day', y='Tip', hue='Gender', dodge=True)
sns.swarmplot(data=tips, x='Day', y='Tip', hue='Gender', dodge=True, color='black', marker='*')
fig.tight_layout()
```

```python
fig = plt.figure(figsize=(6,6))
sns.violinplot(data=tips, x='Day', y='Tip', hue='Gender', dodge=True, palette='pastel', split=True)
fig.tight_layout()
```

```python
fig = plt.figure(figsize=(6,6))
sns.violinplot(data=tips, x='Day', y='Total_bill', hue='Gender', dodge=True, palette='pastel', split=True)
fig.tight_layout()
```

```python
fig = plt.figure(figsize=(6,6))
sns.catplot(data=tips, x='Day', y='Total_bill', hue='Gender', kind='swarm')
fig.tight_layout()
```

```python
fig = plt.figure(figsize=(6,6))
sns.catplot(data=tips, x='Day', y='Total_bill', hue='Gender', kind='box', col='Time')
fig.tight_layout()
```
