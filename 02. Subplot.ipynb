```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

### Subplot

`plt.subplot(fila, columna, Ã­ndice)`

```python
# Subplot me permite plotear varios grÃ¡ficos en un mismo lienzo

# Figura
plt.figure(figsize=(7,3))

# Subplot 1
plt.subplot(1,3,1)
plt.plot(x,y,'r--')
plt.grid()
plt.title('Line plot')

# Subplot 2
plt.subplot(1,3,2)
plt.hist(x,color='orange')
plt.grid()
plt.title('Hist plot')

# Subplot 3
plt.subplot(1,3,3)
plt.boxplot(x)
plt.grid()
plt.title('Box plot')

# Visualizar
plt.tight_layout()
```

```python
# Subplot me permite plotear varios grÃ¡ficos en un mismo lienzo

# Figura
plt.figure(figsize=(7,3))

# Subplot 1
plt.subplot(1,2,1)
plt.plot(x,y,'o-')
plt.plot(y,x,'k:')
plt.grid()
plt.title('Line plot')

# Subplot 3
plt.subplot(1,2,2)
plt.pie(y)
plt.grid()
plt.title('Pie plot')

# Visualizar
plt.tight_layout()
```
