```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
# Figura
fig = plt.figure()

# Ejes
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9])
axes.plot(x, y, 'b')

# Visualizar
plt.show()
```

```python
# Figura
fig = plt.figure()

# Ejes
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9])
axes2 = fig.add_axes([0.11, 0.73, 0.3, 0.25])

axes.plot(x, y, 'b')
axes2.plot(y, x, 'b')
axes2.set_facecolor('yellow')

# Visualizar
plt.show()
```
