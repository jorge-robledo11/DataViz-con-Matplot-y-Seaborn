```python
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.simplefilter("ignore")
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, ((ax1, ax2)) = plt.subplots(nrows=1, ncols=2, figsize=(8,4))

ax1.plot(x, y, label='$Sin(x)$', c='r', marker='*', ls='-')
ax1.set_title('Relación X - Y')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.legend()
ax1.grid()

ax2.plot(y, x, label='$Sin(x)$', c='b', marker='d', ls=':')
ax2.set_title('Relación Y - X')
ax2.set_xlabel('Y')
ax2.set_ylabel('X')
ax2.legend()
ax2.grid()

fig.tight_layout()
```

```python
plt.plot(x, y, label='Sin(x)')
plt.title('Título')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```
