```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')
tips = tips.rename(columns={'sex': 'gender'})
tips = tips.rename(str.capitalize, axis='columns')
tips
```

```python
sns.histplot(data=tips, x='Tip', hue='Gender', stat='count', multiple='dodge')
plt.tight_layout()
```

```python
sns.kdeplot(data=tips, x='Tip', hue='Gender', shade=True)
plt.tight_layout()
```

```python
sns.displot(data=tips, x='Tip', hue='Gender', kind='kde')
plt.tight_layout()
```

```python
sns.displot(data=tips, x='Tip', hue='Gender', kind='hist', multiple='stack')
plt.tight_layout()
```

```python
fig, axes = plt.subplots(figsize=(8,6))
sns.kdeplot(data=tips, x='Tip', hue='Day', shade=True)
fig.suptitle('Density Plot', fontsize=16, horizontalalignment='center')
sns.set(style="dark", palette="muted", font="Segoe UI")
sns.color_palette("flare", as_cmap=True)
plt.grid()
fig.tight_layout()
```
