```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings
warnings.simplefilter("ignore")

tips = sns.load_dataset('tips')
tips = tips.rename(columns={'sex': 'gender'})
tips = tips.rename(str.capitalize, axis='columns')
tips
```

```python
fig = plt.figure(figsize=(6,6))
sns.jointplot(data=tips, x='Total_bill', y='Tip', hue='Gender', palette='dark')
fig.tight_layout()
```

```python
fig = plt.figure(figsize=(6,6))
sns.jointplot(data=tips, x='Total_bill', y='Tip', hue='Gender', palette='dark', kind='kde', marginal_ticks=True)
fig.tight_layout()
```

```python
fig = plt.figure(figsize=(6,6))
sns.jointplot(data=tips, x='Total_bill', y='Tip', hue='Gender', palette='dark', kind='hist', marginal_ticks=True)
fig.tight_layout()
```

```python
# Cruza todas las variables numÃ©ricas
fig = plt.figure(figsize=(6,6))
sns.pairplot(data=tips, hue='Gender', palette='coolwarm')
fig.tight_layout()
```
