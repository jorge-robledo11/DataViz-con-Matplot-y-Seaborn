```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings
warnings.simplefilter("ignore")

tips = sns.load_dataset('tips')
tips = tips.rename(columns={'sex': 'gender'})
tips = tips.rename(str.capitalize, axis='columns')
tips
```

```python
tips.groupby('Day').size()
```

```python
fig = plt.figure(figsize=(6,6))
sns.scatterplot(data=tips, x='Total_bill', y='Tip', hue='Day', style='Time', alpha=0.75, size='Size')
fig.tight_layout()
```

```python
# Podemos customizar los markers en un diccionario
markers = {'Lunch': 'D', 'Dinner': 's'}
fig = plt.figure(figsize=(6,6))
sns.scatterplot(data=tips, x='Total_bill', y='Tip', hue='Day', style='Time', alpha=0.75, size='Size', markers=markers)
fig.tight_layout()
```

```python
# Podemos customizar los markers en un diccionario
markers = {'Lunch': 'D', 'Dinner': 's'}
fig = plt.figure(figsize=(6,6))
sns.lineplot(data=tips, x='Total_bill', y='Tip', hue='Day', style='Time', alpha=0.75, size='Size', markers=markers)
fig.tight_layout()
```

```python
# Podemos customizar los markers en un diccionario
markers = {'Lunch': 'D', 'Dinner': 's'}
fig = plt.figure(figsize=(6,6))
sns.relplot(data=tips, x='Total_bill', y='Tip', hue='Day', style='Time', alpha=0.75, size='Size', markers=markers)
fig.tight_layout()
```

```python
# Podemos customizar los markers en un diccionario
markers = {'Lunch': 'D', 'Dinner': 's'}
fig = plt.figure(figsize=(6,6))
sns.relplot(data=tips, x='Total_bill', y='Tip', hue='Day', style='Time', alpha=0.75, size='Size', markers=markers, kind='line')
fig.tight_layout()
```
