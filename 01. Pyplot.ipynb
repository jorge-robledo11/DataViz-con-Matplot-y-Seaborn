```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
array = np.linspace(0,5,11)
y = array**2
```

```python
plt.plot(array, y)
plt.tight_layout()
```

```python
plt.plot(array, y, 'rs-')
plt.tight_layout()
```

```python
plt.hist(array)
plt.tight_layout()
```

```python
plt.pie(array)
plt.tight_layout()
```

```python
plt.scatter(array, y)
plt.tight_layout()
```

```python
plt.boxplot(array)
plt.tight_layout()
```
