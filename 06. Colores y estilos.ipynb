```python
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.simplefilter("ignore")
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
for i in plt.style.available:
  plt.style.use(i)
  fig, ax = plt.subplots(figsize=(3,2))
  ax.plot(x, y)
  ax.set_title(i)
```

```python
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, x+1, color='green', alpha=1, lw=1, ls='-', marker='o', markersize=7)
ax.plot(x, x+2, color='yellow', alpha=1, lw=1, ls='--', marker='d', markersize=7)
ax.plot(x, x+3, color='orange', alpha=1, lw=1, ls=':', marker='*', markersize=10)
ax.plot(x, x+4, color='white', alpha=1, lw=1, ls='dashed', marker='^', markersize=7)
ax.set_title('Gráfico')
ax.set_yticks(np.arange(0, 11, 2))
fig.tight_layout()
```
