```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

tips = sns.load_dataset('tips')
tips
```

```python
tips = tips.rename(columns={'sex': 'gender'})
tips = tips.rename(str.capitalize, axis='columns')
tips
```

```python
sns.displot(data=tips, x='Total_bill')
plt.tight_layout()
```

```python
sns.displot(data=tips, x='Total_bill', y='Tip')
plt.tight_layout()
```

```python
sns.displot(data=tips, x='Total_bill', y='Tip', hue='Gender')
plt.tight_layout()
```

```python
sns.displot(data=tips, x='Total_bill', y='Tip', hue='Gender', kind='kde')
plt.tight_layout()
```

```python
sns.displot(data=tips, x='Total_bill', y='Tip', hue='Gender', kind='kde', palette='dark', alpha=0.75)
plt.tight_layout()
```
