```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings
warnings.simplefilter("ignore")

tips = sns.load_dataset('tips')
tips = tips.rename(columns={'sex': 'gender'})
tips = tips.rename(str.capitalize, axis='columns')
tips
```

```python
tips.corr()
```

```python
fig = plt.figure(figsize=(6,4))
sns.heatmap(data=tips.corr(), annot=True, square=True, cmap='coolwarm', lw=2, vmax=1)
fig.tight_layout()
```
